## Part A – Concept Application (40%)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

print(f'Mean    : {data.mean():.4f}')
print(f'Variance: {data.var():.4f}')
print(f'Std Dev : {data.std():.4f}')

plt.figure(figsize=(8,4))
plt.hist(data, bins=30, edgecolor='white')
plt.title('Normal Distribution (μ=50, σ=10, n=1000)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('am_normal_hist.png', dpi=150)
plt.show()
```

```python
def binomial_pmf(n, p):
    from math import comb
    pmf = {k: comb(n, k) * (p**k) * ((1-p)**(n-k)) for k in range(n+1)}
    print(f'Total probability: {sum(pmf.values()):.6f}')
    return pmf

pmf = binomial_pmf(n=10, p=0.5)
print({k: round(v, 4) for k, v in pmf.items()})
```

```python
def normal_pdf(x, mu, sigma):
    coeff = 1 / (sigma * (2 * np.pi) ** 0.5)
    exponent = np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return coeff * exponent

x_vals = np.linspace(20, 80, 200)
pdf_vals = normal_pdf(x_vals, mu=50, sigma=10)

plt.figure(figsize=(8,4))
plt.plot(x_vals, pdf_vals)
plt.title('Normal PDF (manual)')
plt.xlabel('x')
plt.ylabel('Density')
plt.tight_layout()
plt.savefig('pm_normal_pdf.png', dpi=150)
plt.show()
```

```python
dataset = np.random.normal(loc=60, scale=15, size=500)

mean   = np.mean(dataset)
median = np.median(dataset)
mode   = stats.mode(dataset, keepdims=True).mode[0]
skew   = stats.skew(dataset)

print(f'Mean    : {mean:.2f}')
print(f'Median  : {median:.2f}')
print(f'Mode    : {mode:.2f}')
print(f'Skewness: {skew:.4f} ({"right" if skew > 0 else "left"} skewed)')
```

```python
tosses = np.random.choice(['H', 'T'], size=1000, p=[0.5, 0.5])
p_heads = np.sum(tosses == 'H') / 1000
print(f'Estimated P(heads): {p_heads:.3f}')
print(f'Theoretical P(heads): 0.500')
print(f'Difference: {abs(p_heads - 0.5):.3f}')
```

## Part B – Stretch: Compare Distributions (30%)

```python
np.random.seed(42)
normal_s  = np.random.normal(loc=0, scale=1, size=1000)
uniform_s = np.random.uniform(low=-3, high=3, size=1000)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(normal_s, bins=30, edgecolor='white')
axes[0].set_title(f'Normal  | mean={normal_s.mean():.2f} std={normal_s.std():.2f}')
axes[0].set_xlabel('Value')

axes[1].hist(uniform_s, bins=30, edgecolor='white')
axes[1].set_title(f'Uniform | mean={uniform_s.mean():.2f} std={uniform_s.std():.2f}')
axes[1].set_xlabel('Value')

plt.tight_layout()
plt.savefig('pm_distribution_compare.png', dpi=150)
plt.show()

print('Normal  - bell shaped, most values near mean')
print('Uniform - flat, all values equally likely')
```

## Part C – Interview Ready (20%)

**Q1 — PMF vs PDF**
- PMF: discrete. P(X=k) = exact probability at point k. Values sum to 1.
- PDF: continuous. Probability of interval = area under curve. P(X=k)=0 at a point.

**Q3 — Central Limit Theorem**
As sample size n→∞, distribution of sample means → Normal, regardless of underlying distribution.
- Why important in ML: justifies using Normal-based statistics (t-tests, confidence intervals) on non-normal data; explains why averages of large batches behave predictably.

```python
def empirical_probability(data, event_fn):
    hits = sum(1 for x in data if event_fn(x))
    return hits / len(data)

scores = [45, 72, 88, 55, 63, 91, 39, 76, 82, 50, 94, 60]
p_above_70 = empirical_probability(scores, lambda x: x > 70)
print(f'P(score > 70): {p_above_70:.3f}')
```
