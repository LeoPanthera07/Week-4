## Part A – Concept Application

```python
from functools import reduce
import numpy as np

def mean_loop(data):
    total = 0
    for x in data:
        total += x
    return total / len(data)

def median_loop(data):
    s = sorted(data)
    n = len(s)
    mid = n // 2
    return (s[mid - 1] + s[mid]) / 2 if n % 2 == 0 else s[mid]

def variance_loop(data):
    m = mean_loop(data)
    total = 0
    for x in data:
        total += (x - m) ** 2
    return total / len(data)

sample = [12, 23, 3, 34, 45, 56, 5]
print('Mean:', mean_loop(sample))
print('Median:', median_loop(sample))
print('Variance:', variance_loop(sample))
```

```python
def mean_hof(data):
    return reduce(lambda a, b: a + b, data) / len(data)

def variance_hof(data):
    m = mean_hof(data)
    return reduce(lambda acc, x: acc + (x - m)**2, data, 0) / len(data)

print('Mean (HOF):', mean_hof(sample))
print('Variance (HOF):', variance_hof(sample))
```

```python
student_marks = [45, 72, 88, 55, 63, 91, 39, 76, 82, 50]
avg = mean_loop(student_marks)

above_avg = list(filter(lambda x: x > avg, student_marks))

grades = ['A' if m >= 80 else 'B' if m >= 60 else 'C' for m in student_marks]

print('Average:', avg)
print('Above average:', above_avg)
print('Grades:', grades)
```

```python
def std_dev(data):
    return variance_loop(data) ** 0.5

print('Std Dev (manual):', std_dev(sample))
print('Std Dev (NumPy) :', np.std(sample))
print('Match:', round(std_dev(sample), 8) == round(np.std(sample), 8))
```

```python
A = [12, 23, 3, 34, 45, 56, 5]
B = [12, 1, 3, 1, 1, 2, 3, 4, 5, 3, 4]

mean_A, var_A = mean_loop(A), variance_loop(A)
mean_B, var_B = mean_loop(B), variance_loop(B)

print(f'A: mean={mean_A:.2f}, var={var_A:.2f}')
print(f'B: mean={mean_B:.2f}, var={var_B:.2f}')
print(f'Difference in means: {mean_A - mean_B:.2f}')
```

## Part B – Stretch: Two-Sample t-Test 

```python
def two_sample_t_test(s1, s2, alpha=0.05):
    n1, n2 = len(s1), len(s2)
    m1, m2 = mean_loop(s1), mean_loop(s2)
    v1, v2 = variance_loop(s1), variance_loop(s2)
    t_stat = (m1 - m2) / ((v1/n1 + v2/n2) ** 0.5)
    print(f't-statistic: {t_stat:.4f}')
    print('Reject H0' if abs(t_stat) > 1.96 else 'Fail to reject H0')
    return t_stat

two_sample_t_test(A, B)
```

**Q3**: Hypothesis testing checks if data supports a claim.
- H₀ (null): no effect
- p-value: probability of result under H₀
- α = 0.05: threshold below which we reject H₀

Example: Testing if new drug reduces BP — H₀: no difference. p=0.02 < 0.05 → reject H₀.

```python
def process_list(nested):
    flat = [x for sublist in nested for x in sublist]
    odds_only = [x for x in flat if x % 2 != 0]
    return flat, odds_only

nested = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat, odds = process_list(nested)
print('Flat:', flat)
print('Odds only:', odds)
```
